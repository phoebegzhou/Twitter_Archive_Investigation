```python
#Query twitter API for additional information
import time
import json
count=0
start=time.time()
fails_dict={}
with open('tweet_json.txt','w') as outfile: 
    for line in df.tweet_id:
        count=count+1
        print(str(count)+': '+str(line))
        
        try: 
            tweet = api.get_status(line, tweet_mode='extended')
            #convert to json and write to the outfile
            print('success')
            json.dump(tweet._json,outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print('fail')
            fails_dict[line]=e
            pass
end=time.time()
print(end-start)
print(fails_dict)    
```

`twitter-archive-enhanced` table: rating_denominator has values other than 10.

##### Define:
Locate all rows with rating_denominator that is not 10. Find all "number/number" pattern from text using regex for all rows that rating_denominator is not 10. Join all number/number values together by comma using apply function. Create a new column called 'rating' and assign the second number/number value to it because by reading the text, the second number/number values are correct ratings. Split the rating column on '/' to obtain 'rating_1' and 'rating_2' columns. Assign values from rating_1 column to rating_numerator if not null. Assign values from rating_2 column to rating_denominator if not null. Drop rating, rating_1, and rating_2 columns. 

```python
mask=df_clean.rating_denominator !=10
df_clean['rating']=df_clean[mask].text.str.findall('(\d+/\d+)').apply(','.join).str.split(',', 1).str[1]
df_clean['rating_1']=df_clean['rating'].str.split("/").str[0]
df_clean['rating_2']=df_clean['rating'].str.split("/").str[1]
df_clean['rating_numerator']=df_clean['rating_1'].fillna(df_clean['rating_numerator'])
df_clean['rating_denominator']=df_clean['rating_2'].fillna(df_clean['rating_denominator'])
df_clean=df_clean.drop(columns=['rating','rating_1','rating_2'],axis=1)
```

`twitter-archive-enhanced` table: Real names after 'named' in text need to be extracted and replace the wrong names.

##### Define:
Replace "name" column that has values "None" with "none". Locate rows with name that does not start with upper case letter (name start with upper case letter). Look at the texts to find out real names are after the word "named". Find all "named". Extract the first word after "named" and put them into a new column called 'real_name'. Replace the real name with original name if not null. Drop the 'real_name' column.

##### Code:

```python
df_clean['name'].replace("None",'none',inplace=True)
mask2=df_clean.name.str[0].str.isupper().fillna(False)
df_clean['real_name']=df_clean[~mask2].text.str.findall('named\s\w+').apply(' '.join).str.split(' ',1).str[1]
df_clean['name']=df_clean['real_name'].fillna(df_clean['name'])
df_clean.drop('real_name',axis=1, inplace=True)
```

##### Test:

`twitter-archive-enhanced` table: some rows are not about dog ratings but for other animals.

##### Define:
Find the rows with text that contains word 'only rate' and put them into df_new. Filter the rows with "none" value and put them into df_new1. Find the rows that contain upper case letter (names start with upper case letters) and put them to df_new2. Append df_new2 to df_clean. Drop df_new rows.

##### Code:

```python
df_new=df_clean[df_clean['text'].str.contains('only rate')]
df_new1=df_new[df_new.name != "none"]
df_new2=df_new1[df_new1.name.str.contains('[A-Z]')]
df_clean=df_clean.append(df_new2)
df_clean.drop(df_new.index, inplace=True)
```

##### Tidiness issue:

Doggo, floofer, pupper, puppo in `twitter-archive-enhanced` table should be variables under 'stage' column.

##### Define: 
Import Numpy. Replace "None" in the doggo, floofer, pupper, puppo columns to "". Create a new column called 'stage' to store doggo, floofer, pupper, puppo. Assign multiple stages to variables that contain two stages. Assign NaN value back to variables that do not contain any of the four stages. Drop doggo, floofer, pupper, puppo columns. 

##### Code:

```python
import numpy as np
df_clean.doggo.replace("None","",inplace=True)
df_clean.floofer.replace("None","",inplace=True)
df_clean.pupper.replace("None","",inplace=True)
df_clean.puppo.replace("None","",inplace=True)
df_clean['stage'] = df_clean.doggo + df_clean.floofer + df_clean.pupper + df_clean.puppo

df_clean.loc[df_clean.stage == 'doggopupper', 'stage'] = 'doggo,pupper'
df_clean.loc[df_clean.stage == 'doggopuppo', 'stage'] = 'doggo,puppo'
df_clean.loc[df_clean.stage == 'doggofloofer', 'stage'] = 'doggo,floofer'
df_clean.loc[df_clean.stage == '', 'stage'] = np.nan
df_clean=df_clean.drop(['doggo','floofer','pupper','puppo'],axis=1)
```

##### Test:

```python
df_clean.stage.value_counts()
```

tweet_id column is duplicated in all three tables.

##### Define:
Convert tweet_id in df3_clean into int64. Merge df_clean, df2_clean, and df3_clean into one table called df_merge, joining on tweet_id.

##### Code:

```python
df3_clean['tweet_id']=df3_clean['tweet_id'].astype(np.int64)
df_merge=pd.merge(df_clean,df2_clean, on='tweet_id')
df_merge=pd.merge(df_merge,df3_clean, on='tweet_id')
```

### Part IV. Data Analysis and Visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline 
df_master=pd.read_csv('twitter_archive_master.csv')
df_master
```

```python
#Create a column called "ratings", which equals to rating_numerator divided by rating_denominator. 
df_master['rating']=df_master['rating_numerator']/df_master['rating_denominator']

#Calculate the mean rating for each dog stage: doggo, pupper, puppo, and floofer.
doggo_mean_rating=df_master[df_master['stage'].astype('str').str.contains('doggo')]['rating'].mean()
pupper_mean_rating=df_master[df_master['stage'].astype('str').str.contains('pupper')]['rating'].mean()
puppo_mean_rating=df_master[df_master['stage'].astype('str').str.contains('puppo')]['rating'].mean()
floofer_mean_rating=df_master[df_master['stage'].astype('str').str.contains('floofer')]['rating'].mean()
doggo_mean_rating, pupper_mean_rating, puppo_mean_rating, floofer_mean_rating
```

```python
#Create a bar char for mean rating for each dog stage
locations=[1,2,3,4]
heights=[doggo_mean_rating, pupper_mean_rating, puppo_mean_rating, floofer_mean_rating]
labels=['doggo','pupper','puppo','floofer']
plt.bar(locations,heights,tick_label=labels)
plt.title('Average Rating by Dog Stage')
plt.xlabel('Dog Stage')
plt.ylabel('Average Rating');
```

**Conclusion:**
From the bar chart above, we can see that when a dog is in "puppo" stage, it gets the highest average rating of 1.20. When a dog is in "pupper" stage, it gets the lowest average rating of 1.07.

```python
#Calculate the average favorite counts for each dog stage
doggo_mean_favorite_count=df_master[df_master['stage'].astype('str').str.contains('doggo')]['favorite_count'].mean()
pupper_mean_favorite_count=df_master[df_master['stage'].astype('str').str.contains('pupper')]['favorite_count'].mean()
puppo_mean_favorite_count=df_master[df_master['stage'].astype('str').str.contains('puppo')]['favorite_count'].mean()
floofer_mean_favorite_count=df_master[df_master['stage'].astype('str').str.contains('floofer')]['favorite_count'].mean()
doggo_mean_favorite_count, pupper_mean_favorite_count, puppo_mean_favorite_count, floofer_mean_favorite_count
```

```python
#Create a bar char for average favorite counts for each dog stage
locations=[1,2,3,4]
heights=[doggo_mean_favorite_count, pupper_mean_favorite_count, puppo_mean_favorite_count, floofer_mean_favorite_count]
labels=['doggo','pupper','puppo','floofer']
plt.bar(locations,heights,tick_label=labels)
plt.title('Average Favorite Counts by Dog Stage')
plt.xlabel('Dog Stage')
plt.ylabel('Average Favorite Counts');
```

**Conclusion:**
From the bar chart above, we can see that "puppo" stage has the highest average favorite counts, "doggo" stage has the second highest average favorite counts, and "pupper" stage has the lowest average favorite counts.

```python
#Create a new column called "dog_breed". Loop through dog predictions and return corresponding dog breed if True. 
def predict_breed(x):
    if x.p1_dog:
        return x.p1
    elif x.p2_dog:
        return x.p2
    elif x.p3_dog:
        return x.p3
df_master['dog_breed'] = df_master.apply(lambda x: predict_breed(x),axis=1)       
```
